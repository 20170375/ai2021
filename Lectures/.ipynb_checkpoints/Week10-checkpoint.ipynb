### 10주차 오픈소스 프레임워크를 이용한 고급 인공신경망의 구현

#### 1. 텐서플로우로 시작하는 Hello world! - MNIST 예제

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt

mnist = tf.keras.datasets.mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()

print('x_train 데이터의 형태:', x_train.shape)
print('x_train[0] 데이터의 형태:', x_train[0].shape)
print('y_train 데이터의 형태:', y_train.shape)
print('x_test 데이터의 형태:', x_test.shape)
```

```python
num = x_train[0]
for i in range(28):
    for j in range(28):
        print('{:4d}'.format(num[i][j]), end='')
    print()
```

```python
plt.imshow(num, cmap='Greys', interpolation='nearest')
```

5라고 레이블링 되어있다.

```python
print('y_train[0] =', y_train[0])
```

#### 2. keras로 순차 심층 신경망 구축하기

compile() 함수에서 모델의 하이퍼파라미터로

+ 최적화 함수를 무엇으로 할 것인지를 설정하는 optimizer
+ 손실 함수는 어떤 것을 사용할지를 설정하는 loss
+ 훈련기간동안 어떠한 측정 도구를 사용하여 정답 레이블과의 차이를
측정할 지를 결정하는 metrics 등이 있음

```python
x_train, x_test = x_train / 255, x_test / 255 # 입력값 정규화

model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(256, activation='relu'),
    keras.layers.Dense(10, activation='softmax'),
])

# 학습을 위한 최적화 함수, 손실 함수등을 가진 모델을 컴파일
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=5)
```

```python
model.summary()
```

학습에 사용하지 않은 1만개의 숫자 이미지가 있는 x_test를 이 모델에 넣어서 y_test와 비교

```python
print('신경망 모델의 학습 결과')
eval_loss, eval_acc = model.evaluate(x_test, y_test)
print('test 데이터의 손실값', eval_loss, 'test 데이터의 정확도', eval_acc)
```

#### 3. 정규화와 표준화

정규화<sup>normalization</sup>
  
> $ \tilde{x} = {{x - x_{min}} \over {x_{max} - x_{min}}} $


표준화<sup>standardization</sup> 

> $ x' = {{x - \mu_x} \over {\sigma_x}} $

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

yr      = np.array([15,30,23,5,9,43,33,29,3,56])
n_rooms = np.array([2,4,3,4,4,3,3,1,1,2])

min_max_scaler = MinMaxScaler() # 정규화
yr_norm      = min_max_scaler.fit_transform(yr[:, np.newaxis])
n_rooms_norm = min_max_scaler.fit_transform(n_rooms[:, np.newaxis])

plt.figure(figsize=(12, 4))
plt.subplot(121)
plt.scatter(yr, n_rooms, c='red')
plt.xlim(0, 50)
plt.ylim(0, 50)
plt.xlabel('year')
plt.ylabel('num of rooms')

plt.subplot(122)
plt.scatter(yr_norm, n_rooms_norm, c='red')
plt.xlim(-0.1, 1.1)
plt.ylim(-0.1, 1.1)
plt.xlabel('year')
plt.ylabel('num of rooms')
```

#### 4. 경사 하강법 변형

**배치 경사 하강법**<sup>batch gradient descent</sup>
+ 전체 데이터 셋을 모두 사용
+ 데이터 집합이 너무 클 경우 **최적해를 얻는데 너무 많은 계산 시간이 소
요**된다는 **단점이 있음**

**확률적 경사 하강법**<sup>stochastic gradient descent</sup>
+ 무작위로 뽑은 하나의 훈련 데이터에 대하여 경사 하강법을 적용하는 방법
+ 실행 속도가 빠르지만 **무작위로 뽑은 데이터의 편향**으로 인하여 신경망의 성능이 들쑥날쑥 변하며 학습


**미니 배치 경사 하강법**<sup>mini-batch gradient descent</sup>
+ 배치 경사 하강법과 확률적 경사 하강법 간의 절충안
+ 무작위로 선택한 n개의 데이터를 사용하여 신경망을 학습시키는 방법으
로 학습 파라미터가 많은 경우에도 n개의 데이터만을 샘플링하므로 해를
구하는 계산 시간이 절약되며, 여러 개의 데이터를 학습에 사용하므로 데
이터의 편향도 비교적 적다는 장점

#### 5. 더 깊은 층으로 정확도를 높여보고 교차검증도 하자

```python
# 층을 추가한 모델 만들기
model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(10, activation='softmax'),
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))
```

```python
# 더 많은 층을 가진 신경망 모델 만들기
model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(10, activation='softmax'),
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))
```

이전 모델에 비해 신경망 층의 수도 대폭 늘고, 학습해야 할 파라미터도
크게 늘었지만, 학습 데이터와 검증 데이터에 대해 측정된 정확도가 거의
변화가 없음
+ 신경망을 더 깊게 만드는 것만으로는 인식률 높이기에 한계가
있음을 보이는 것

```python
import pandas as pd
pd.DataFrame(history.history).plot(figsize=(5, 3))
```

#### 6. 학습데이터에만 최적화된 신경망 개선하기 : 드롭아웃

드롭아웃<sup>dropout</sup> 혹은 희석화라고도 하는 이 방법은 오버피팅
<sup>overfitting</sup> 문제를 해결하는 대안
+ 이 방법의 핵심 아이디어는 그림과 같이 신경망의 학습 단계
에서 임의의 노드를 탈락<sup>drop-out</sup>시켜 학습시키는 것

케라스 API에서는 손쉽게 드롭아웃 레이어를 생성하는 방법을 제공

```python
data = np.arange(1, 11).reshape(2, 5).astype(np.float32)
print(data)

layer = tf.keras.layers.Dropout(0.3, input_shape=(2, ))
outputs = layer(data, training=True)
print(outputs)

# 매번 정확하게 30%만 탈락시키는 것이 아니라 30%의 확률로 탈락이 이루어짐
```

```python
# 드롭아웃 계층을 가진 신경망 모델 만들기
model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(10, activation='softmax'),
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))
```

```python
pd.DataFrame(history.history).plot(figsize=(5, 3))
```

+ 드롭아웃 노드를 포함한 신경망 모델의 학습과정을 그려 보면 학습 데이터에 대해 정확도가 더 천천히 개선되고 있음
+ 검증 데이터에 대해서 드롭아웃이 없는 모델에 비해 언제나 좋은 예측 성능을 보이고 있어 과적합 문제를 잘 회피하며 학습을 진행하고 있음
