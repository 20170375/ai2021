## 7. SVM - ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞ Î®∏Ïã†<sup>support vector machine</sup>Ïùò ÏÜåÍ∞ú
SVMÏùÄ Îëê Îç∞Ïù¥ÌÑ∞ Í∑∏Î£πÏùÑ ÎÇòÎàÑÎäî Ï¥àÌèâÎ©¥ÏùÑ Ï∞æÏúºÎ©¥ÏÑú Ìè≠<sup>margin</sup>Ïù¥ Í∞ÄÏû• ÎÑìÏùÄ Í≤ÉÏùÑ Ï∞æÎäî Î∞©Î≤ï

## 8. ÌïòÎìú ÎßàÏßÑ ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞ Î®∏Ïã†Ïùò Íµ¨ÌòÑ

ÏµúÏ†ÅÌôî: $ \underset{w,b}{argmin}  {1 \over 2} w^T w $

Ï†úÏïΩÏ°∞Í±¥: i=1,2,...,mÏù∏ mÍ∞ú Îç∞Ïù¥ÌÑ∞ Ïù∏Ïä§ÌÑ¥Ïä§ $ x^{(i)} $ Î™®ÎëêÏóê $ | w^T x^{(i)} + b | \ge 1 $


## 9. ÏÜåÌîÑÌä∏ ÎßàÏßÑ ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞ Î®∏Ïã†Ïùò Íµ¨ÌòÑ

ÏµúÏ†ÅÌôî: $ \underset{w,b}{argmin}  {1 \over 2} w^T w + C \sum_{i=1}^m \zeta^{(i)} $

Ï†úÏïΩÏ°∞Í±¥: mÍ∞ú Îç∞Ïù¥ÌÑ∞ Ïù∏Ïä§ÌÑ¥Ïä§ $ x^{(i)} $ Î™®ÎëêÏóê ÎåÄÌï¥ $ | w^T x^{(i)} + b | \ge 1 - \zeta^{(i)} $

<br>

ÏõêÎûòÏùò ÏµúÏ†ÅÌôî Î¨∏Ï†úÏòÄÎçò $ {w^ùëáw} $ Î•º ÏµúÏÜåÎ°ú ÎßåÎìúÎäî Î¨∏Ï†úÏôÄ Ìï®Íªò $ ùõ¥_{i=1}^ùëö \zeta^{(i)} $ Î•º ÏµúÏÜåÌôîÌïòÎäî Î¨∏Ï†úÎ•º ÎèôÏãúÏóê ÌíÄÏñ¥Ïïº Ìï®.

Ï†úÏñ¥ Î≥ÄÏàò CÍ∞Ä 0Ïù¥ÎùºÎ©¥ Ïä¨Îûô Î≥ÄÏàòÏóê ÎåÄÌïú ÏµúÏ†ÅÌôîÍ∞Ä Ï†ÑÌòÄ Ïù¥Î£®Ïñ¥ÏßÄÏßÄ ÏïäÏùÑ Í≤ÉÏù¥Î©∞, Ïù¥ Í∞íÏùÑ ÌÅ¨Í≤å ÎëêÎ©¥ Ïä¨Îûô Î≥ÄÏàòÎ•º ÏµúÏÜåÌôîÌïòÎäî ÏùºÏóê ÎπÑÏ§ëÏù¥ Îçî Ïª§ÏßÄÍ≤å Îê† Í≤É


## 10. ÏÇ¨Ïù¥ÌÇ∑Îü∞ÏùÑ Ïù¥Ïö©Ìïú ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞ Î®∏Ïã† ÏÇ¨Ïö©ÌïòÍ∏∞

```python
import pandas as pd
import numpy as np
data_loc = 'https://github.com/dknife/ML/raw/main/data/'
df = pd.read_csv(data_loc + 'two_classes.csv')
df.tail(5)
```

```python
df_positive = df[df['y']  > 0] # yÍ∞Ä 1Ïù∏ Îç∞Ïù¥ÌÑ∞Îßå Ï∂îÏ∂ú
df_negative = df[df['y'] == 0] # yÍ∞Ä 0Ïù∏ Îç∞Ïù¥ÌÑ∞Îßå Ï∂îÏ∂ú
import matplotlib.pyplot as plt
plt.scatter(df_positive['x1'], df_positive['x2'], color='r')
plt.scatter(df_negative['x1'], df_negative['x2'], color='g')
```

ÏÇ¨Ïù¥ÌÇ∑Îü∞ÏùÑ Ïù¥Ïö©ÌïòÏó¨ Ïã§Ï†úÎ°ú ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞ Î®∏Ïã†ÏùÑ Ïù¥Ïö©ÌïòÏó¨ Îç∞Ïù¥ÌÑ∞Î•º Íµ¨Î∂Ñ

```python
from sklearn.svm import LinearSVC
X = df[['x1', 'x2']].to_numpy()           # x1, x2Î•º ÏûÖÎ†• Î≤°ÌÑ∞Î°ú ÌïúÎã§
y = df['y']                               # yÏó¥Ïùò Í∞íÏù¥ Î†àÏù¥Î∏î
svm_simple = LinearSVC(C=1, loss='hinge') # SVM ÌÅ¥ÎûòÏä§ ÏÉùÏÑ± (Í≤ΩÏ≤©hinge ÏÜêÏã§ Ìï®Ïàò)
svm_simple.fit(X, y)                      # ÏûÖÎ†•Í≥º Î†àÏù¥Î∏îÎ°ú SVM ÌïôÏäµ Ïã§Ïãú

# Î†àÏù¥Î∏î ÏòàÏ∏°
svm_simple.predict([[0.12, 0.56], [-4, 40], [0, 40], [5, 20]])
```

SVM Î∂ÑÎ•òÍ∏∞Ïùò Î∂ÑÎ¶¨ Í≤ΩÍ≥ÑÎ•º ÌôïÏù∏ÌïòÍ≥† Ïã∂ÏúºÎ©¥ Îã§ÏùåÍ≥º Í∞ôÏù¥ ÌïôÏäµÎêú svm_simpleÎ°ú DecisionViz ÌÅ¥ÎûòÏä§Î•º ÎßåÎì§Í≥† fit()Í≥º draw()

```python
from yellowbrick.contrib.classifier import DecisionViz
viz = DecisionViz(svm_simple, title="linear SVM")
viz.fit(X, y)
viz.draw(X, y)
```

## 11. ÌååÏù¥ÌîÑÎùºÏù∏ÏùÑ Ïù¥Ïö©Ìïú Îç∞Ïù¥ÌÑ∞ Ï†ïÏ†ú

SVMÏùÄ Îç∞Ïù¥ÌÑ∞Ïùò ÌÅ¨Í∏∞Ïóê ÎØºÍ∞êÌïú ÌäπÏÑ±ÏùÑ Í∞ñÍ≥† ÏûàÏùå

Îç∞Ïù¥ÌÑ∞ Ï†ïÏ†úÎ•º ÏúÑÌï¥ **Ï†ïÍ∑úÌôî**<sup>normalization</sup> ÌòπÏùÄ **ÌëúÏ§ÄÌôî**<sup>standardization</sup>ÏùÑ ÏàòÌñâÌï®

#### ÏÇ¨Ïù¥ÌÇ∑ Îü∞ Ìå®ÌÇ§ÏßÄÎäî Ïù¥Î•º ÏúÑÌï¥ **ÌååÏù¥ÌîÑÎùºÏù∏**<sup>pipeline</sup>Ïù¥ÎùºÎäî Ìé∏Î¶¨Ìïú Î∞©Î≤ïÏùÑ Ï†úÍ≥µ

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC

svm_std = Pipeline([                       # ÌååÏù¥ÌîÑÎùºÏù∏ÏúºÎ°ú SVM Í∞ùÏ≤¥ Íµ¨ÏÑ±
  ("std", StandardScaler()),               # Îç∞Ïù¥ÌÑ∞ ÌëúÏ§ÄÌôî Îã®Í≥Ñ Ìè¨Ìï®
  ("lsvm", LinearSVC(C=1, loss='hinge')),  # ÏÑ†Ìòï SVM Î∂ÑÎ•òÍ∏∞ Ìè¨Ìï®
])
svm_std.fit(X, y)
```

## 12. Îã§Ìï≠ ÌäπÏßï Î≥ÄÌôòÏùÑ ÌÜµÌïú ÎπÑÏÑ†Ìòï ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞ Î®∏Ïã†Ïùò Íµ¨ÌòÑ

Îã§Ìï≠ ÌöåÍ∑ÄÏóê ÏÇ¨Ïö©ÌñàÎçò PolynomialFeaturesÎäî Îç∞Ïù¥ÌÑ∞ Ï†ïÏ†ú Îã®Í≥ÑÏù¥ÎØÄÎ°ú ÌïôÏäµ ÏïåÍ≥†Î¶¨Ï¶òÏóê Í¥ÄÍ≥Ñ ÏóÜÏù¥ Ïì∏ Ïàò ÏûàÏùå

Îç∞Ïù¥ÌÑ∞ Ï†ïÏ†úÎ•º Ìè¨Ìï®Ìïú ÌïôÏäµÎ™®Îç∏ÏùÑ ÎßåÎì§Í∏∞ ÏúÑÌï¥ ÏÇ¨Ïö©ÌñàÎçò **ÌååÏù¥ÌîÑÎùºÏù∏**<sup>pipeline</sup>Ïóê ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞Ïùò Îã§Ìï≠ ÌäπÏßï Î≥ÄÌôòÏùÑ Ìè¨Ìï®ÌïòÎäî Í≤É

```python
from sklearn.preprocessing import PolynomialFeatures
svm_poly = Pipeline([
  ("std", StandardScaler()),
  ("poly_inputs", PolynomialFeatures(degree=5)),
  ("lsmv", LinearSVC(C=0.01, loss='hinge'))
])
svm_poly.fit(X, y)
viz = DecisionViz(svm_poly, title="polynomial feature SVM")
viz.fit(X, y)
viz.draw(X, y)
```
