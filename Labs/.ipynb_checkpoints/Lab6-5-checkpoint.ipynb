## LAB<sup>6-5</sup> 커널 트릭을 이용한 비선형 SVM

**실습목표**

선형 분리가 불가능하도록 섞여있는 두 그룹의 데이터가 아래에서 있다.

https://github.com/dknife/ML/raw/main/data/twisted_data.csv

비선형 SVM으로 이 데이터들을 분류하되 다항 특징 변환을 사용하지 않고 해결하라.

### 1. 이미 사용한 데이터이므로 아래와 같이 같은 방식으로 읽고, 화면에 표시하고, 데이터를 준비하는 과정을 동일하게 수행한다.

```python
import pandas as pd
import numpy as np

data_loc = 'https://github.com/dknife/ML/raw/main/data/'
df = pd.read_csv(data_loc + 'twisted_data.csv')

from sklearn.svm import SVC # 커널 트릭을 사용할 수 있는 SVM 분류기
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from yellowbrick.contrib.classifier import DecisionViz

X = df[['x1', 'x2']].to_numpy()
y = df['y']
```

### 2. 파이프라인을 이용하여 데이터 표준화를 거친 데이터를 SVM 분류기에 넘기도록 한다. 이때 커널 트릭이 가능한 SVC로 생성해야 하며, 이 클래스의 생성자에는 loss 키워드 매개변수가 없다. 그리고 커널을 지정할 수 있도록 하는데, 다항 커널의 경우 'poly'를 kernel 키워드 매개변수에 지정한다. 다항 커널의 차수도 지정하며,  coef0는 커널에 있는 독립항 r을 의미한다.

```python
polynomial_svm_clf = Pipeline([
  ("scaler", StandardScaler()),
  ("svm_clf", SVC(C=10, kernel='poly', degree=4, coef0=2))
])
polynomial_svm_clf.fit(X, y)
viz = DecisionViz(polynomial_svm_clf, title="polynomial feature SVM")
viz.fit(X, y)
viz.draw(X, y)
```
