## LAB<sup>6-1</sup> 다항 회귀의 회귀 함수를 그려 보자.

**실습목표**

아래의 위치에는 직선으로 표현하기 어려운 데이터가 있다.

https://github.com/dknife/ML/raw/main/data/nonlinear.csv

이 데이터를 화면에 그려보고 이 데이터를 설명하는 회귀 함수를 다항 회귀를 이용하여 찾아 보자. 그리고 회귀 함수를 데이터의 독립 변수 범위 내에서 가시화해 보라.

1. 필요한 패키지를 임포트하고, 데이터를 읽어 온다. 그리고 데이터를 그려보자.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
data_loc = 'https://github.com/dknife/ML/raw/main/data/'
df = pd.read_csv(data_loc + 'nonlinear.csv')

plt.scatter(df['x'], df['y'])
```

2. 데이터를 3차 다항 회귀를 위해 변환한다.

```python
from sklearn.preprocessing import PolynomialFeatures
X = df['x'].to_numpy()
y = df['y'].to_numpy()
X = X.reshape(-1,1) # 입력을 2차원 벡터가 되게 한다. shape = (m, 1)
feature_cubic = PolynomialFeatures(degree=3)
X_3 = feature_cubic.fit_transform(X)
```

3. 선형회귀 모델을 만들고, 회귀 함수를 그릴 때 사용할 독립변수 범위를 0에서 1까지로 한다. 이때 입력은 2차원 배열이 되어야 하므로, reshape(-1,1)을 지정한다. -1은 데이터의 크기에 따라 적절한 값으로 맞추라는 의미이며, 두 번째 차원은 1이 되도록 하였다. 데이터 인스턴스가 m개라면 shape = (m,1)의 배열을 만들게 된다.

```python
lin_model = LinearRegression()
domain = np.linspace(0, 1, 100).reshape(-1,1) # 입력은 2차원 벡터로 변형
```

4. 3차 다항 회귀를 위한 입력과 출력 레이블의 쌍으로 학습을 실행하고, 이 모델을 이용하여 회귀 곡선을 그려 보자.

```python
lin_model.fit(X_3, y)
domain_3 = feature_cubic.fit_transform(domain)
predictions = lin_model.predict(domain_3)
plt.scatter(df['x'], df['y'])
plt.scatter(domain, predictions, color='r')
```
