## LAB<sup>7-4</sup> 입력 다항화로 XOR를 해결해 보기 (p.248)

**실습 목표**

참 또는 거짓의 값을 갖는 두 개의 입력에 대해 XOR 연산을 수행할 수 있도록, 입력에 대한 다항 변환을 실시하고, 이를 통해 얻은 다항 특징에 대해 퍼셉트론 학습을 통해 XOR 연산을 수행하는 모델을 구현하자.

### 1. 두 입력에 대한 다항 변환으로 위의 그림과 같은 특징을 얻고, 여기에 퍼셉트론 학습을 적용한다면 모두 4개의 원소를 가진 가중치 벡터가 필요하다. 학습률도 설정한다.

```python
import numpy as np

W = np.array([0,0,0,0])
learning_rate = 0.01
```

### 2. 활성화 함수는 동일한 계단함수를 사용하자.

```python
def activation(s):
  if s > 0:   return 1
  elif s < 0: return -1
  return 0
```

### 3. 네트워크의 출력은 다항 변환이 이루어진 입력에 대해 네 개의 원소를 가진 가중치 벡터가 곱해진다. 구현 자체는 앞에서 살펴본 단순한 퍼셉트론과 동일하지만, 입력으로 사용되는 것이 다항 특징임을 명확히 하기 위해 매개변수 이름을 polyX로 변경해 보자.

```python
def out(polyX):
  return activation(W.dot(polyX))
```

### 4. 학습은 입력에 대해 다항 변환을 실시하는 일을 포함하여 진핸된다. 제일 먼저 입력을 변환하여 polyX를 생성하고, 이를 out() 함수에 넘겨 출력 값을 계산한다. 퍼셉트론 학습은 다항 입력 polyX와 목표값 target의 곱을 이용하여 표현된다.

```python
def train(x0, x1, target):
  global W

  polyX = np.array([x0, x1, x0*x1, 1])
  y     = out(polyX)

  ### 예측이 맞으면 아무것도 하지 않음-------------------------------------
  if target == y: return 0                # 가중치가 변경되지 않았음을 반환
  ### 예측이 틀리면 학습 실시----------------------------------------------
  print('가중치 수정전 target : {} y:{}, W:{}'.format(target, y, W))
  W = W + learning_rate * polyX * target  # 입력x목표값 비례하여 변경
  print('가중치 수정후 target : {} y:{}, W:{}'.format(target, y, W))
  return 1                                # 가중치가 변경되었음을 반환
```

### 5. 입력에 대한 예측 역시 각 입력 데이터를 다항화한 뒤에 out() 함수를 불러 이루어진다.

```python
def predict(inputs):
  outputs = []
  for x in inputs:
    polyX = np.array([x[0], x[1], x[0]*x[1], 1])
    outputs.append (out(polyX))
  return outputs
```

### 6. 훈련과 예측은 앞에서 수행한 것과 같은 방식으로 하면 된다. XOR 연산의 입력과 목표 출력을 주고 학습을 수행하고, 얻은 모델에 입력을 제공하면 바른 동작을 확인할 수 있다.

```python
adjusted = 0
for i in range(100):
  adjusted += train(-1,-1, -1)    # 훈련 데이터 1
  adjusted += train(-1, 1,  1)    # 훈련 데이터 2
  adjusted += train( 1,-1,  1)    # 훈련 데이터 3
  adjusted += train( 1, 1, -1)    # 훈련 데이터 4
  print("iteration ------------", i)
  if not adjusted: break    # 모든 훈련에 대해 가중치 변화 없으면 학습종료
  adjusted = 0

X = [[-1,-1], [-1,1], [1,-1], [1,1]]
yhat = predict(X)
print('x0 x1  y')
for i in range(len(X)):
  print('{0:2d} {1:2d} {2:2d}'.format(X[i][0], X[i][1], yhat[i]))
```
