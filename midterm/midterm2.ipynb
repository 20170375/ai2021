### 문제 2

```python
# 2-1
import numpy as np
from skimage.io import imread
from skimage.transform import resize

url = './data/detector/faces/'

imgR, imgC, channel = 64, 64, 3
face_images = []

for i in range(100):
  file = url + 'img00{0:03d}.jpg'.format(i+1) # 이미지 파일 이름을 만든다.
  img = imread(file)                          # 이미지를 읽는다.
  img = resize(img, (imgR, imgC ,channel))    # 이미지의 크기를 변경한다.
  face_images.append(img)                          # images 리스트에 담는다.
    
face_images = np.array(face_images)

### 이미지 데이터를 증강시키기 ###
from keras.preprocessing.image import ImageDataGenerator
image_generator = ImageDataGenerator(
    rotation_range=10,
    zoom_range=0.1,
    shear_range=0.1,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    vertical_flip=False
)

face_images = face_images.reshape(-1, imgR, imgC, channel)
it = image_generator.flow(face_images)

# 데이터 증강
n_augmentation = 16
for _ in range(n_augmentation):
    new_image = it.next()
    face_images = np.append(face_images, new_image)
    
face_images = np.array(face_images*255, dtype=np.uint)
face_images = face_images.reshape(-1, imgR, imgC, channel)

# 총 개수 확인
face_images = face_images[:500]
print("face_images.shape : ", face_images.shape)
```

```python
# 2-2
from skimage.feature import hog         # Histogram of Gradient 계산
face_hogs     = []
face_features = []

for i in range(500):
    hog_desc, hog_image = hog(face_images[i], orientations=8,
                            pixels_per_cell=(16,16), cells_per_block=(1,1),
                            visualize=True, multichannel=True)
    face_hogs.append(hog_image)
    face_features.append(hog_desc)
```

```python
# 2-3
import matplotlib.pyplot as plt

def plot_images(nRow, nCol, img):
  fig = plt.figure()
  fig, ax = plt.subplots(nRow, nCol, figsize = (nCol,nRow))
  for i in range(nRow):
    for j in range(nCol):
      if nRow <= 1: axis = ax[j]
      else:         axis = ax[i, j]
      axis.get_xaxis().set_visible(False)
      axis.get_yaxis().set_visible(False)
      axis.imshow(img[i*nCol+j])

plot_images(10, 10, face_images)
plot_images(10, 10, face_hogs)
fig = plt.figure()
fig, ax = plt.subplots(10, 10, figsize=(10,6))
for i in range(10):
  for j in range(10):
    ax[i, j].imshow(resize(face_features[i*10 + j], (128,16)))
```

```python
# 2-4
url = './data/detector/animals/'

animal_images = []

for i in range(100):
  file = url + 'img00{0:03d}.jpg'.format(i+1) # 이미지 파일 이름을 만든다.
  img = imread(file)                          # 이미지를 읽는다.
  img = resize(img, (imgR, imgC ,channel))    # 이미지의 크기를 변경한다.
  animal_images.append(img)                          # images 리스트에 담는다.
    
animal_images = np.array(animal_images)

### 이미지 데이터를 증강시키기 ###
image_generator = ImageDataGenerator(
    rotation_range=10,
    zoom_range=0.1,
    shear_range=0.1,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    vertical_flip=False
)

animal_images = animal_images.reshape(-1, imgR, imgC, channel)
it = image_generator.flow(animal_images)

# 데이터 증강
n_augmentation = 16
for _ in range(n_augmentation):
    new_image = it.next()
    animal_images = np.append(animal_images, new_image)
    
animal_images = np.array(animal_images*255, dtype=np.uint)
animal_images = animal_images.reshape(-1, imgR, imgC, channel)

# 총 개수 확인
animal_images = animal_images[:500]
print("animal_images.shape : ", animal_images.shape)

animal_hogs     = []
animal_features = []

for i in range(500):
    hog_desc, hog_image = hog(animal_images[i], orientations=8,
                            pixels_per_cell=(16,16), cells_per_block=(1,1),
                            visualize=True, multichannel=True)
    animal_hogs.append(hog_image)
    animal_features.append(hog_desc)

def plot_images(nRow, nCol, img):
  fig = plt.figure()
  fig, ax = plt.subplots(nRow, nCol, figsize = (nCol,nRow))
  for i in range(nRow):
    for j in range(nCol):
      if nRow <= 1: axis = ax[j]
      else:         axis = ax[i, j]
      axis.get_xaxis().set_visible(False)
      axis.get_yaxis().set_visible(False)
      axis.imshow(img[i*nCol+j])

plot_images(10, 10, animal_images)
plot_images(10, 10, animal_hogs)
fig = plt.figure()
fig, ax = plt.subplots(10, 10, figsize=(10,6))
for i in range(10):
  for j in range(10):
    ax[i, j].imshow(resize(animal_features[i*10 + j], (128,16)))
```

```python
# 2-5
X, y = [], []

for feature in face_features:
  X.append(feature)
  y.append(1)
for feature in animal_features:
  X.append(feature)
  y.append(0)

X = np.array(X)
y = np.array(y)

print("X.shape :", X.shape)
print("y.shape :", y.shape)
```

```python
# 2-6
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

polynomial_svm_clf = Pipeline([
    ("scaler", StandardScaler()),
    ("svm_clf", SVC(C=1, kernel='rbf', gamma=0.01))
 ])
polynomial_svm_clf.fit(X, y)
```
