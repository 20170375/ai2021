### 문제 1

```python
# 1-1
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("./data/nonlinear.csv");

x = data['x']
y = data['y']

plt.scatter(x, y)
```

```python
# 1-2
x = data['x']
x = x[:, np.newaxis]
X = np.c_[np.ones((len(x), 1)), x]

What = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

def ypred(x, What):
  return x*What[1] + What[0]

plt.scatter(x, y)
plt.plot([0, 1], [ypred(0, What), ypred(1, What)], color='r')
```

```python
# 1-3
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
poly_feature = PolynomialFeatures(degree=4)
X = poly_feature.fit_transform(X)

X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.2)
lin_model = LinearRegression()
lin_model.fit(X_train, y_train)

y_hat_train = lin_model.predict(X_train)
y_hat_test = lin_model.predict(X_test)
plt.scatter(y_train, y_hat_train, color='r')
plt.scatter(y_test, y_hat_test, color='b')
```
